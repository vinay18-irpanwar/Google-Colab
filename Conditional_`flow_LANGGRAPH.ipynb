```python
from typing_extensions import TypedDict
#add state

class Returnstate(TypedDict):
  days_since_delivery:str
  result:str
```

```python
from langgraph.graph import StateGraph,START,END

workflow=StateGraph(Returnstate)
```

```python
#add node

def check_refund_policy(state:Returnstate):
  days=int(input("enter how many days since delivery:"))
  state["days_since_delivery"]=days
  return state
def full_refund(state:Returnstate):
    state["result"]="full refund"
    return state

def partial_refund(state:Returnstate):
    state["result"]="partial refund"
    return state
def no_refund(state:Returnstate):
    state["result"]="no refund"
    return state


workflow.add_node("check_refund_policy_node",check_refund_policy)
workflow.add_node("full_refund_node",full_refund)
workflow.add_node("partial_refund_node",partial_refund)
workflow.add_node("no_refund_node",no_refund)
```

```python
#add edges

def repeater(state:Returnstate):
  if state["days_since_delivery"]>=30:
    return "no_refund"
  elif state["days_since_delivery"]<=7:
    return "full_refund"
  else:
    return "partial_refund"
workflow.add_edge(START,"check_refund_policy_node")
workflow.add_conditional_edges("check_refund_policy_node",repeater,
         {"partial_refund":"partial_refund_node",
         "full_refund":"full_refund_node",
         "no_refund":"no_refund_node"}
         )
workflow.add_edge("partial_refund_node",END)
workflow.add_edge("full_refund_node",END)
workflow.add_edge("no_refund_node",END)
```

```python
app=workflow.compile()
```

```python
res=app.invoke({})
print(res)
```
