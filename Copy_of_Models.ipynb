#Prompting tool using OpenAI with Python

```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('openai_temp'))
response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {"role": "user", "content": "Give me 5 famous tourist places in Banglore."}
    ]
)
print(response.choices[0].message.content)
```

# Prompting tool using Gemini with Python

```python
from google import genai

client = genai.Client(api_key=userdata.get('GOOGLE'))

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Give me 5 famous tourist places in Banglore.",
)

print(response.text)
```

```python
from langchain_openai import ChatOpenAI


llm = ChatOpenAI(model="gpt-4",openai_api_key=userdata.get('openai_temp'))

response = llm.invoke("Give me 5 famous tourist places in Banglore.")

print(response.content)
```
