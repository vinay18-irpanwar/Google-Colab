##step 1 => uploading raw documents
Using
* colab file


```python
from google.colab import files

upload=files.upload()
file_name=list(upload.keys())[0]
```

*  Extracting text using Pymupdf
**Document loader**

```python
!pip install pymupdf
```

```python
from langchain_community.document_loaders import PyMuPDFLoader

content=PyMuPDFLoader(file_name).load()
print(content)
```

```python
content1=[doc.page_content for doc in content]
content="".join(content1)
print(content)
```

##step 2 => Converting raw text into Chunks
Using
* text splitter

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

chunks=RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=0).split_text(content)
print(chunks)
```

##step 3 => creating embeddings and store in vector stors

Using
* FAISS
* googlegenerativeaiembeddings

Note
* vector stores => store only document

```python
from langchain_core.documents import Document
from google.colab import userdata
#
doc_chunks= [Document(page_content=text) for text in chunks]
print(doc_chunks)
```

FAISS

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS

vector_store=FAISS.from_documents(
    embedding=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",google_api_key=userdata.get('lang')),
    documents=doc_chunks,
)
```

```python
query=input("enter quetion related rules in cricket ")
```

```python
res=vector_store.similarity_search_with_score(query=query,k=4)
for doc in res:
   context=doc[0].page_content
```

LLM


```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
LLM=ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key=userdata.get('lang'))

prompt=PromptTemplate(template="act as a cricket expert and response to user query:-{query} and refer given context:-{context}",input=["query","context"])

chain=prompt | LLM | StrOutputParser()

ress=chain.invoke({"query":query,"context":context})
print(ress)
```
