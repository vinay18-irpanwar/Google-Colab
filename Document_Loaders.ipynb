# Student task

```python
from langchain_community.document_loaders import CSVLoader

from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
```

```python
loader = CSVLoader(filename)

docs = loader.load()

print(docs[0].page_content)
print(docs[1].page_content)
```

```python
full_csv_content = "\n".join([row.page_content for row in docs])

print(full_csv_content)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata


api_key = userdata.get("gea_7")

llm = ChatOpenAI(
    model="gpt-4",
    openai_api_key=api_key
)

prompt = PromptTemplate(
    template="""
You are an HR assistant. You are given structured CSV data about applicants.

Answer the user's question based only on the following data:
----------------
{content}
----------------
User's Question: {question}
""",
    input_variables=["content", "question"]
)

applicant_chain = prompt | llm | StrOutputParser()


que = "list frontend candidates with experience > 3 years"
res = applicant_chain.invoke({"content":full_csv_content,"question":que})


print(res)
```
