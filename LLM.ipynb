```python
from openai import OpenAI

from google.colab import userdata

client = OpenAI(api_key=userdata.get('mykey'))

response = client.chat.completions.create(
    model="gpt-4",
    messages= [
        {"role":"user","content":"Who is PM of India"}
    ]
)

print(response.choices[0].message.content)
```

# Chatbot

```python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get("GOOGLE_API_KEY"))


#list =[userinput, AI response]
history = []
while True:
  user_input = input("Enter your prompt")
  if user_input == "exit":
    break

  history.append(user_input)
  print(history,"current_history")
  response = client.models.generate_content(
      model="gemini-2.5-flash",
      contents = history
  )

  ai_response = (response.text)
  print(ai_response)
  history.append(ai_response)
```

```python
from google import genai

client = genai.Client(api_key=userdata.get("GOOGLE_API_KEY"))

user_input = input("enter a topic of your choice")

response = client.models.generate_content(
    model="gemini-1.5-flash",
    contents = f"Write a detailed notes on {user_input}"
)

content = response.text

print(content)
```

```python
# 2nd step: content => LLM => response

interview_content = client.models.generate_content(
    model="gemini-1.5-flash",
    contents = f"Give me 5 important questions on {content}"
)

print(interview_content.text)

```
