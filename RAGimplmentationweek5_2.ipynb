```python
raw_data = """

Types of Leaves

- Casual: 8 days/year. Can be used for personal matters. Must be applied at least 1 day in advance.

- Sick: 10 days/year. Can be availed with a medical certificate if more than 2 days.

- Earned: 15 days/year. Must be pre-approved by your manager.

- Maternity: 26 weeks for female employees, as per government norms.

- Paternity: 7 days for new fathers, to be taken within 2 months of childbirth.

- Bereavement: 3 days, in case of immediate family loss.

"""
```

##step 2

```python
chunked_data=raw_data.split("\n")
print(chunked_data)
```

###step4

```python
matches=[]
def search(user_query,chunked_data):
  words=user_query.split( )
  for i in chunked_data:
    for word in words:
      if word.lower() in i.lower():
         matches.append(i)
  return matches
user_query=input("enter text")
search_result=search(user_query,chunked_data)
```

```python
from google import genai
from google.colab import userdata

client=genai.Client(api_key=userdata.get('chaning'))


```

```python
prompt=f"give me info on {user_query} and refere this {search_result}"
def llm(prompt):
  response=client.models.generate_content(
      model="gemini-2.5-flash",
      contents=prompt,
  )
  print(response.text)
llm(prompt)
```
