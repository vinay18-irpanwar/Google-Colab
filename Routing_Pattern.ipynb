```python
from google import genai

from google.colab import userdata
client = genai.Client(api_key=userdata.get('GOOGLE'))



#Agent
def get_routed_agent(user_input):
  routing_prompt = f"""Based on {user_input}, select one agent from the following agents
  1. return_agent
  2. refund_agent
  3. general_support_agent.

  Return me only name of the agent. For example: return_agent

  """
  response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=routing_prompt,
)

  return response.text.lower()



```

```python
#Return Agent

def agent_return(user_input):
  return_prompt = f"""

    You are a specialized "Return Agent" for an e-commerce company. Your sole responsibility is to assist users with product returns. You are helpful, polite, and precise.
    Whenever user asks any query : {user_input} around returning products, answer him politely that your query has been accepted and it will be resolved soon

  """
  response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=return_prompt,
)

  return response.text


```

```python
#Refund Agent

def agent_refund(user_input):
  refund_prompt = f"""

    You are a specialized "Refund Agent" for an e-commerce company. Your sole responsibility is to assist users with refunding amount . You are helpful, polite, and precise.
    Whenever user asks any query : {user_input} around refund , answer him politely that your query has been accepted and it will be resolved soon.

  """
  response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=refund_prompt,
)

  return response.text


```

```python
#General support Agent

def agent_general_support(user_input):
  general_prompt = f"""

    You are a specialized "General support Agent" for an e-commerce company. Your sole responsibility is to assist users with any general queries  . You are helpful, polite, and precise.
    Whenever user asks any query : {user_input} around any general queries , answer him politely that your query has been accepted and it will be resolved soon.

  """
  response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=general_prompt,
)

  return response.text


```

```python
AGENT_FUNCTIONS_MAPPING = {
    "return_agent": agent_return,
    "refund_agent":agent_refund,
    "general_support_agent":agent_general_support
}
```

```python


user_input = "What is my return status of my Iphone"


def runner(user_input):
  selected_agent = get_routed_agent(user_input)
  print("-----------Selected Agent---------", selected_agent)
  agent_function = AGENT_FUNCTIONS_MAPPING.get(selected_agent)
  res = agent_function(user_input)
  print(res)


runner(user_input)
```
