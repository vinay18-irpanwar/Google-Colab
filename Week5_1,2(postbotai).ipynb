```python



def web_search(user):
  combine=""
  response = client.search(
    query=user
  )

  for res in response["results"]:
      combine=combine+res.get("content")
  return combine







```

```python

def summarize_agent(search_results_content):
    summary_prompt = f"Summarize the following research content: {search_results_content} into a short paragraph:\n"
    response = gemini_client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=summary_prompt
    )
    return (response.text)



```

```python
def linkdin_agent(summary):
    post_content = f"""
                 you are a prosonal branding expert for linkdin.
                 write a professional, insight and engaging linkdin post based on
                 summary below .the tone should be thoughtful and value-driven.avoid hashtags
                 user request{summary}
                 guidelines:
                 -write in the first person .
                 -keep it within 2-4 short paragraphs.
                 -Aim to educate,inspire or provoke thought."""
    response = gemini_client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=post_content
    )
    return (response.text)


linkdin= summarize_agent(summary)
```

```python
def fack_book_agent(summary):
    post_content = f"yoy are a witty and creative facebook caption writer. user reuest  {summary}/n]"
    response = gemini_client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=post_content
    )
    return (response.text)


fack_book= summarize_agent(summary)
```

```python
def twitter_agent(summary):
    post_content = f"Generate a twitter post based on {summary}/n]"
    response = gemini_client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=post_content
    )
    return (response.text)


twitter= summarize_agent(summary)
```
