# ***LLM***

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key=userdata.get('gemikey'))
```

## **STATE**

```python
from typing import TypedDict
from langgraph.graph.message import add_messages
from typing_extensions import Annotated
class ChatBotstate(TypedDict):
  message:Annotated[list,add_messages]
```

```python
from langgraph.graph import StateGraph
workflow=StateGraph(ChatBotstate)
```

## **add nodes**

```python
def chatnode(state:ChatBotstate):
  current_message=state["message"]
  return {"message":[llm.invoke(current_message)]}


workflow.add_node("chatnode",chatnode)
```

## **add edges**

```python
from langgraph.graph import START,END
workflow.add_edge(START,"chatnode")
workflow.add_edge("chatnode", END)
```

```python
app=workflow.compile()
```

## **invoke**

```python
while(True):
  query=input("YOU:")
  if query.strip().lower() in ["exit","by"]:
    break

  res=app.invoke({"message":query})
  #["message"][0].content
  print(res)
```
