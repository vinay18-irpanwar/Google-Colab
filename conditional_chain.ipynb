```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

LLM

```python
from google.colab import userdata

LLM=ChatGoogleGenerativeAI(model="gemini-2.5-flash",
                           google_api_key=userdata.get('lang'))
```

message from user

```python
message="hi i am big fan of you"

```

classifys type of message

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain.output_parsers import StructuredOutputParser
message_type=PromptTemplate(template="classify the type of {message} and state that wheather it is query,fan or collabration response should be in one word [query,collab,fan] ",input_varable=["message"])

message_chain=message_type | LLM | StrOutputParser()
```

collab,query and fan prompt and respective chain

```python
query_prompt=PromptTemplate(template="Write a friendly and helpful reply on a {message}",input=["message"])
collab_prompt=PromptTemplate(template="Write a professional yet warm reply to a {message} request.",input=["message"])
fan_prompt=PromptTemplate(template="Write a heartfelt and positive reply to a  {message}",input=["message"])


query_chain=query_prompt | LLM | StrOutputParser()
collab_chain=collab_prompt | LLM | StrOutputParser()
fan_chain=fan_prompt | LLM | StrOutputParser()
```

for conditional chain we use runnable branch

```python
from langchain_core.runnables import RunnableBranch,RunnableLambda

select_chain=RunnableBranch(
    (lambda x:x.strip().lower() == "query",query_chain),
    (lambda x:x.strip().lower() == "collab",collab_chain),
    (lambda x:x.strip().lower() == "fan",fan_chain),
    RunnableLambda(lambda x:"not match")


)


final_chain=message_chain | select_chain
res=final_chain.invoke({"message":message})
print(res)
```
