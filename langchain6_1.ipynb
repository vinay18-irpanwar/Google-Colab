creating LLM using gemini in python


```python
from google import genai
from google.colab import userdata
api_key = userdata.get('gemikey')


client = genai.Client(api_key=api_key)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="top 5 place in bangluru",
)

print(response.text)
```

creating LLL using langchain (MODELS)

```python
from langchain_google_genai import ChatGoogleGenerativeAI   #for text generation we use chatgooglegenerativeai
from google.colab import userdata

llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key=api_key)
response=llm.invoke("top 5 place in bangluru")

print(response.content)
```

creating llm using langchain gemini for embedding models

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",google_api_key=api_key)
response=embeddings.embed_query("top 5 place in bangluru")

print(response)
```
