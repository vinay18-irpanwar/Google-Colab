```python
from typing_extensions import TypedDict
#state

class Tempstate(TypedDict):
  celsius:float   #-> input
  fahrenheit:float   # -> output
  kelvin:float     # -> output
```

```python
#add to graph

from langgraph.graph import StateGraph,START,END

workflow=StateGraph(Tempstate)
```

```python
#add node
def celcious(state:Tempstate):
  c=float(input('enter temp:') )
  state["celsius"]=c
  print(f"recived temprature in celsius {state["celsius"]} C")
  return state
def kelvin(state:Tempstate):
  celsius=state["celsius"]
  kelvin=celsius+273.1
  state["kelvin"]=kelvin
  return {"kelvin":kelvin}
def fahrenheit(state:Tempstate):
  celsius=state["celsius"]
  fahrenheit=(celsius*9/5)+32
  state["fahrenheit"]=fahrenheit
  return {"fahrenheit":fahrenheit}
workflow.add_node("input_temprature_node",celcious)
workflow.add_node("kelvin_node",kelvin)
workflow.add_node("fahrenheit_node",fahrenheit)
```

```python
workflow.add_edge(START,"input_temprature_node")
workflow.add_edge("input_temprature_node","kelvin_node")
workflow.add_edge("input_temprature_node","fahrenheit_node")


workflow.add_edge("kelvin_node",END)
workflow.add_edge("fahrenheit_node",END)

```

```python
app=workflow.compile()
```

```python
#invocation of graph

#message={"celsius":75}
# USER INPUT PASS EMPTY INITIAL STATE
res=app.invoke({})
print(res)
```
