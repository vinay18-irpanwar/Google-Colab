```python
from langchain_core.documents import Document
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma
from google.colab import userdata
```

use local server / store localy in memory

```python
docs1=Document(page_content="hi my name is vinay")
docs2=Document(page_content="hi my name is yxz")
docs=[docs1,docs2]
embed=GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001",
    google_api_key=userdata.get('lang')
)
```

```python
vector=Chroma(
    embedding_function=embed,
    persist_directory="db",
    collection_name="my_collection"
)

vector.add_documents(docs)
```

```python
res=vector.similarity_search("vinay")
print(res)
```
